# Q. Create a code to simulate a game of Rock Paper Scissors.
 
You should be able to simulate:
- Human vs Human
- Human vs PC

```python
from random import choice

class RockPaperScissors:
    def __init__(self):
        self.WIN_MAP = {
            "rock": "scissors",
            "scissors": "paper",
            "paper": "rock"
        }
    #selection the game mode i.e user vs user or user vs computer
    def select_game_mode(self):
        print("""Select the game mode:
            --> Enter 1 for Human vs Human
            --> Enter 2 for Human vs Computer""")
        game_mode = input("Enter the game mode you want: ")
        if game_mode == "1":
            return 1
        elif game_mode == "2":
            return 2
        else:
            print("Invalid Choice!! Try again with entries 1 or 2")
            self.select_game_mode()
    #logic to determine the winner between two choices given
    def determine_winner(self, player_choice, another_player_choice):
        if player_choice == another_player_choice:
            print("It's a Tie.")
        elif self.WIN_MAP[player_choice] == another_player_choice:
            print(f"The winner is {player_choice}")
        else:
            print(f"The winner is {another_player_choice}")
    #code to get the users choices and if choosen, computer random choices
    def get_choices(self, game_mode):
        #asking user1 for choice until the valid input is given
        while True:
            user1_choice = input("Enter the choice User1: ").lower()
            if user1_choice not in list(self.WIN_MAP.keys()):
                print(f"User1, you can only have choices from list: {list(self.WIN_MAP.keys())}. Proceed with choice again.")
                continue
            else:
                print(f"User1, use chose: {user1_choice}")
                break

        #asking user2 for choice until the valid input is given
        while game_mode == 1:
            user2_choice = input("Enter the choice User2: ").lower()
            if user2_choice not in list(self.WIN_MAP.keys()):
                print(f"User2, you can only have choices from list: {list(self.WIN_MAP.keys())}. Proceed with choice again.")
                continue
            else:
                print(f"User2, you chose: {user2_choice}")
                return user1_choice, user2_choice
        #below piece of code runs if user1 vs computer is choosen to play
        computer_choice = choice(list(self.WIN_MAP.keys()))
        print(f"Computer chose: {computer_choice}")
        return user1_choice, computer_choice
        
    #main code to call the required functions 
    def start(self):
        print("Let's begin the game:")
        print("------------------------------------------------------")
        #calling function that chooses human vs human or human vs computer
        game_mode = self.select_game_mode()
        #calling function that gets the choises of either users or user and computer
        choice1, choice2 = self.get_choices(game_mode)
        #calling function that inforces logic to determine user and print information
        self.determine_winner(choice1, choice2)


game = RockPaperScissors()
game.start()
```

# Traditional way using List Data Type:

```python
class RockPaperScissors:
    def __init__(self):
        self.choice = ["rock", "paper", "scissors"]

    def get_user_choice(self):
        user_choice = input("Enter your choice: ").lower()
        while user_choice not in self.choice:
            print(f"Invalid choice. Please enter the choices from list:{self.choice}: ")
            user_choice = input("Enter your choice: ")
        return user_choice
    
    def determine_winner(self, choice1, choice2):
        if choice1 == choice2:
            print("It's a choice, run the game again.")
        elif (choice1=="rock" and choice2=="paper") or(choice1=="paper" and choice2=="scissors") or (choice1=="scissors" and choice2=="rock"):
            print("Winner is User1")
        else:
            print("Winner is user2.")

    def play(self):
        print("Let's break DEADLOCK:")
        choice1 = self.get_user_choice()
        choice2 = input("User2 please enter your choice: ").lower()
        print(f"Choice of user1: {choice1} and choice of user2: {choice2}")
        result = self.determine_winner(choice1, choice2)
        print(result)

game = RockPaperScissors()
game.play()
```
